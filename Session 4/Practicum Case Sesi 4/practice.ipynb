```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("picture.jpg", 0)
cv2.imshow("image", img)
cv2.waitKey(0)
```

# Canny

```python
gx = cv2.Sobel(img, cv2.CV_16S, 1, 0)
gy = cv2.Sobel(img, cv2.CV_16S, 0, 1)
```

```python
res = cv2.Canny(img, 200, 100)
resized = cv2.resize(res, (1600, 1000))
cv2.imshow("canny", resized)
cv2.waitKey(0)
```

# Sobel

```python
plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 1)
plt.imshow(gx, 'gray')
plt.title("Sobel X")
plt.subplot(1, 2, 2)
plt.imshow(gy, 'gray')
plt.title("Sobel Y")
plt.show()
```

# Laplace

```python
res = cv2.Laplacian(img, 200, 100)
resized = cv2.resize(res, (1600, 1000))
cv2.imshow("laplace", resized)
cv2.waitKey(0)
```
